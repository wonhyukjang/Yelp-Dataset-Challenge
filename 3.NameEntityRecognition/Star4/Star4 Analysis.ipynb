```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
resultNeg4 = pd.read_csv('star4_resultNegDic.csv')
resultPos4 = pd.read_csv('star4_resultPosDic.csv')
resultNegReason4 = pd.read_csv('star4_resultNegReasonDic.csv')
resultPosReason4 = pd.read_csv('star4_resultPosReasonDic.csv')
```

```python
print 'Before delete 0 occurence', len(resultNeg4)
print 'After delete 0 occurence', len(resultNeg4[resultNeg4['count'] > 0])
resultNegReduced4 = resultNeg4[resultNeg4['count'] > 0]
resultNegReduced4 = resultNegReduced4.sort(['count'],ascending = False)
print resultNegReduced4.describe()
print 'Neg word Sum: ',resultNegReduced4['count'].sum()

word_counts = resultNegReduced4[:50]
resultNegReduced4.set_index(["word"],inplace=True)
resultNegReduced4[:50].plot(kind='bar')
```
