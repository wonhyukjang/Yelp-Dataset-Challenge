```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from collections import OrderedDict
# star1_list = OrderedDict([('pos_word_sum', 1075279), ('neg_word_sum', 134970), ('pos_reason_sum', 2590606), ('neg_reason_sum', 310446)])
# star2_list = OrderedDict([('pos_word_sum', 1075279), ('neg_word_sum', 134970), ('pos_reason_sum', 2590606), ('neg_reason_sum', 310446)])

data = OrderedDict((('pos_word_sum', {'star1': 1075279, 'star2': 1083169,'star3': 1503857, 'star4': 2746129,'star5':2603540}),
                                ('neg_word_sum', {'star1': 134970, 'star2': 90916,'star3': 70103, 'star4': 65322,'star5':46599}),
                                ('pos_reason_sum', {'star1': 2590606, 'star2': 2573011,'star3': 3564779, 'star4': 6381696,'star5':5892882}),
                                ('neg_reason_sum', {'star1': 310446, 'star2': 216844,'star3': 177279, 'star4': 174492,'star5':122837})))
df = pd.DataFrame.from_dict(data)
df['word_sentiment'] = df['pos_word_sum'] - df['neg_word_sum']
df['reason_count'] = df['pos_reason_sum'] - df['neg_reason_sum']
df['word_pos_percentage'] = df['word_sentiment'] / (df['pos_word_sum'] + df['neg_word_sum']) * 100.0
df['reason_pos_percentage'] = df['reason_count'] / (df['pos_reason_sum'] + df['neg_reason_sum']) * 100.0
print df.index.tolist
ind = [1,2,3,4,5]
print ind
plt.plot(ind,df['reason_count'])
# mylist = {('pos_word_sum': 1075279), ('neg_word_sum': 134970), ('pos_reason_sum': 2590606), ('neg_reason_sum': 310446)}
#print star1_list
df.head()
```

```python
plt.plot(ind,df['word_sentiment'])
```

```python
word_counts = df['word_pos_percentage']
df.set_index(df.index,inplace=True)
word_counts.plot(kind='bar', title= 'Positive Word Percentage')
```

```python
word_counts2 = df['reason_pos_percentage']
df.set_index(df.index,inplace=True)
df.columns.name = 'Count'
word_counts2.plot(kind='bar', title= 'Positive Reason Percentage')
```
