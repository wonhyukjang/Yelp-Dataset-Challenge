```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from collections import OrderedDict
# star1_list = OrderedDict([('pos_word_sum', 1075279), ('neg_word_sum', 134970), ('pos_reason_sum', 2590606), ('neg_reason_sum', 310446)])
# star2_list = OrderedDict([('pos_word_sum', 1075279), ('neg_word_sum', 134970), ('pos_reason_sum', 2590606), ('neg_reason_sum', 310446)])

data = OrderedDict((('pos_word_sum', {'star1': 135754, 'star2': 174947,'star3': 325231, 'star4': 822531,'star5':927125}),
                                ('neg_word_sum', {'star1': 1074979, 'star2': 999596,'star3': 1249472, 'star4': 1990593,'star5':1725232}),
                                ('pos_reason_sum', {'star1': 333986, 'star2': 426516,'star3': 777038, 'star4': 1891802,'star5':2061760}),
                                ('neg_reason_sum', {'star1': 2568252, 'star2': 2364443,'star3': 2966752, 'star4': 4668400,'star5':3958833})))
df = pd.DataFrame.from_dict(data)
df['word_sentiment'] = df['neg_word_sum'] - df['pos_word_sum']
df['reason_count'] = df['neg_reason_sum'] - df['pos_reason_sum']
df['word_pos_percentage'] = df['word_sentiment'] / (df['pos_word_sum'] + df['neg_word_sum']) * 100.0
df['reason_pos_percentage'] = df['reason_count'] / (df['pos_reason_sum'] + df['neg_reason_sum']) * 100.0
print df.index.tolist
ind = [1,2,3,4,5]
print ind
plt.plot(ind,df['reason_count'])
# mylist = {('pos_word_sum': 1075279), ('neg_word_sum': 134970), ('pos_reason_sum': 2590606), ('neg_reason_sum': 310446)}
#print star1_list
df.head()
```

```python
word_counts = df['word_pos_percentage']
df.set_index(df.index,inplace=True)
word_counts.plot(kind='bar', title= 'Negative Word Percentage - Positive Word Percentage')
```
