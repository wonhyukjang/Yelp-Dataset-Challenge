```python
resultNeg5 = pd.read_csv('star5_resultNegDic.csv')
resultPos5 = pd.read_csv('star5_resultPosDic.csv')
resultNegReason5 = pd.read_csv('star5_resultNegReasonDic.csv')
resultPosReason5 = pd.read_csv('star5_resultPosReasonDic.csv')
```
