```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
resultNeg3 = pd.read_csv('star3_resultNegDic.csv')
resultPos3 = pd.read_csv('star3_resultPosDic.csv')
resultNegReason3 = pd.read_csv('star3_resultNegReasonDic.csv')
resultPosReason3 = pd.read_csv('star3_resultPosReasonDic.csv')
```

```python
print 'Before delete 0 occurence', len(resultNeg3)
print 'After delete 0 occurence', len(resultNeg3[resultNeg3['count'] > 0])
resultNegReduced3 = resultNeg3[resultNeg3['count'] > 0]
resultNegReduced3 = resultNegReduced3.sort(['count'],ascending = False)
print resultNegReduced3.describe()
print 'Neg word Sum: ',resultNegReduced3['count'].sum()

word_counts = resultNegReduced3[:50]
resultNegReduced3.set_index(["word"],inplace=True)
resultNegReduced3[:50].plot(kind='bar')
```
