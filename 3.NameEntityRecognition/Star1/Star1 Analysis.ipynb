```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
resultNeg = pd.read_csv('star1_resultNegDic.csv')
resultPos = pd.read_csv('star1_resultPosDic.csv')
resultNegReason = pd.read_csv('star1_resultNegReasonDic.csv')
resultPosReason = pd.read_csv('star1_resultPosReasonDic.csv')
```

```python
resultNeg.head()
```

```python
print 'Before delete 0 occurence', len(resultNeg)
print 'After delete 0 occurence', len(resultNeg[resultNeg['count'] > 0])
resultNegReduced = resultNeg[resultNeg['count'] > 0]
resultNegReduced = resultNegReduced.sort(['count'],ascending = False)
resultNegReduced.describe()
print 'Neg word Sum: ',resultNegReduced['count'].sum()

word_counts = resultNegReduced[:50]
resultNegReduced.set_index(["word"],inplace=True)
resultNegReduced[:50].plot(kind='bar')
```
