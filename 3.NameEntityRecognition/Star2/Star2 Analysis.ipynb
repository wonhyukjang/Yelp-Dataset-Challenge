```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
resultNeg2 = pd.read_csv('star2_resultNegDic.csv')
resultPos2 = pd.read_csv('star2_resultPosDic.csv')
resultNegReason2 = pd.read_csv('star2_resultNegReasonDic.csv')
resultPosReason2 = pd.read_csv('star2_resultPosReasonDic.csv')
```

```python
print 'Before delete 0 occurence', len(resultNeg2)
print 'After delete 0 occurence', len(resultNeg2[resultNeg2['count'] > 0])
resultNegReduced2 = resultNeg2[resultNeg2['count'] > 0]
resultNegReduced2 = resultNegReduced2.sort(['count'],ascending = False)
print resultNegReduced2.describe()
print 'Neg word Sum: ',resultNegReduced2['count'].sum()

word_counts = resultNegReduced2[:50]
resultNegReduced2.set_index(["word"],inplace=True)
resultNegReduced2[:50].plot(kind='bar')
```
